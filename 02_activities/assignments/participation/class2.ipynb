```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
```

```python
np.random.seed(613)  # in order to reproduce the same result everytime whoever can run this code
x = np.arange(50)  # generate a 50 numbers from 0 to 49
y1 = np.random.randint(0,100,50) # (min=0, max=100, how many values to generate=50)
y2 = np.random.randint(0,100,50)
# 
```

```python
fig, ax = plt.subplots(figsize=(5,3)) #figsize we gave size of 5X3 inches 
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
fig, ax = plt.subplots(figsize=(5,3)) #figsize we gave size of 5X3 inches 

ax.plot(x,y1, label = "Person 1")  # added label 
ax.plot(x,y2, label = "Person 2")
ax.legend(loc= "lower right") # calling legend to give specific location 
```

```python
# axis labels rotation 
fig, ax = plt.subplots(figsize=(5,3)) #figsize we gave size of 5X3 inches 
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc = "lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# axis roations
plt.xticks(rotation = 45,
           ha = "right",
           fontsize = 12,
           font = "serif")
```
