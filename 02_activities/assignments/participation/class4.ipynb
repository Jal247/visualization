```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
from PIL import Image  # to open image
import requests  # to get images from URL
from io import BytesIO # to store images
```

```python
x= np.array(["Luffy","Zoro","Nami","Usoop", "Sanji"])
y1= np.array([110,180,240, 99, 220])
y2= np.array([170,100,90,120,50])

fig, ax= plt.subplots(figsize=(7,3))

ax.bar(x,y1, color = "indigo")
ax.plot(x,y2,color="red")

```

```python
y2_sd = np.std(y2)

fig, ax= plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = "red")

ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = "none")
```

```python
y2_sd = np.std(y2)

fig, ax= plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = "red")

ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery = 2)
```

```python
# importing an images in to the python 
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color = "red")
# adding a new image on plot
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
# 0.1 is x-coordinate on figure, 0.11 is y-coordinate on figure, image width, image height]
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color = "red")
# adding a new image on plot
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])# axis at x=0.1,y=0.11, height=0.15,tall=0.35

# adding a uploaded image in created plot
ax_image.imshow(image)
ax_image.axis('off')
```
