```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# data related to tips given my customer at restrount
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
sns.set_style('whitegrid') # setting a style of whitegrid

sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
# adding labels using the set.function
tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip')

tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# modify aesthetic elements
fig = plt.subplots(figsize =(10,3))

tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

# adding labels using the set.function

tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')

```

```python
# adding multiple variables to graph
tipgraph = sns.scatterplot(data = tips, 
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time',
                           hue = 'day',
                           palette = ['purple', 'hotpink','deepskyblue', 'yellowgreen'])

# by default it takes lunch and dinner markers
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# Pair plot to compare variables 

sns.pairplot(data = tips,
             hue = 'day')
```

```python
# Relplot also lets us explore specific levels within variables

daysplot = sns.relplot(data = tips,
                      x = "total_bill",
                      y = "tip",
                      hue = "sex",
                      col = "day",
                      kind = "scatter",
                      col_wrap = 2) # definning number of columns like to have

daysplot
```

```python
# plotly 
import plotly.graph_objects as go

x1 = np.array(["luffy", "Zoro", "Nami", "Usoop", "Sanji"])
y1 = np.array([110,180,240,99,220])
```

```python
# plotyly syntax is different then matplotlib
# check features like zoom, box select, autoscale, reset axes, plotly logomark

graph = go.Figure()
graph.add_trace(go.Bar(x = x1, y = y1))

graph.update_layout(title = "Pirate Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score")
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(
    x = x1,
    y = y1,
    mode = 'markers',
    marker = dict(
        size = 15,
        color = 'hotpink',
        opacity = 1,
        line = dict(width = 5, color = 'purple')
    )
    ))

graph.update_layout(title = "Interactive Pirate Plot",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score",
                    width = 500,
                    height = 500)
```

```python
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines = 'skip')
df
```

```python
# make a simple wordcloud using the ''quote' variable

# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)

# generate our wordcloud image
wordcloud = WordCloud(background_color = 'white',
                      colormap = 'inferno').generate(text)

#use matplotlib syntax to put our image in a figure

fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,  # 'imshow' from when we added pictures to our matplotlib axes
          interpolation = 'bilinear' ) # this line helps smooth our image
ax.axis("off")
plt.show()
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(['apple', 'banana','watermelon'])
B = set(['pumkin', 'blueberry', 'apple', 'key lime'])
 
diagram = venn2([A,B],
                 set_labels = ('Fruits', 'Pies'),
                 set_colors = ('blue','red'),
                 alpha = 0.5) # color trasperancy

diagram.get_label_by_id("10").set_text("\n".join(A-B)) # id = 10 means A true(1),B false(0) 
diagram.get_label_by_id("11").set_text("\n".join(A & B)) # id = 11 means A true(1),B true(1)
diagram.get_label_by_id("01").set_text("\n".join(B-A)) # id = 01 means A false(0),B true(1)
```
